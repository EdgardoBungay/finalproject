```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

red = pd.read_csv("winequality-red.csv")
white = pd.read_csv("winequality-white.csv")
red = red.append(white)
wine = red
wine.tail()
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
wine["type"] = encoder.fit(wine["type"]).transform(wine["type"])
wine.head()
# (wine["type"])
# wine.head()
```

```python
X = wine.drop(columns=["type", "quality"])
y = wine["type"]

print(len(X), len(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

X_train.head(1)
print(len(X_train), len(y_train))
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler()
X_test_scaled = X_scaler.fit_transform(X_test)
X_train_scaled = X_scaler.fit_transform(X_train)
```

```python
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)

print(len(X_train_scaled), len(y_train_categorical))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=11))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
# X_train_scaled
```

```python
deep_model = Sequential()
deep_model.add(Dense(units=6, activation='relu', input_dim=11))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
